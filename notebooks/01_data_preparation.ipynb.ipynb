```python
# 1. Import libraries
import pandas as pd
import numpy as np

# 2. Load raw CCHS dataset
df = pd.read_csv("../data/raw/cchs_2022.csv")

print("Raw dataset shape:", df.shape)
```

```python
# 3. Select variables for the project

selected_vars = [
    "LSMDVSWL",
    "DHHGAGE",
    "DHHDGHSZ",
    "GEOGPRV",
    "INCDGHH",
    "EDDVH3",
    "FSCDVHF2",
    "LBFDGWSS",
    "LBFDVPFT",
    "LBFDGHPW",
    "WDMDGDIF",
    "CCCDGSKL",
    "SDCDGIMM",
    "SDCDVFLA"
]

df_selected = df[selected_vars].copy()

print("Selected dataset shape:", df_selected.shape)
df_selected.head()
```

```python
# 4. Save selected dataset

df_selected.to_csv(
    "../data/processed/cchs_selected_variables.csv",
    index=False
)

print("Selected dataset saved.")
```
