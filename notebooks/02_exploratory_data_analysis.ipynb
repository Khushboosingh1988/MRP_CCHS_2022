```python
# ============================================================
# EDA Visualizations
# ============================================================

import os
os.makedirs("../outputs/figures", exist_ok=True)

life_labels = {
    1: "Very Satisfied",
    2: "Satisfied",
    3: "Neither",
    4: "Dissatisfied",
    5: "Very Dissatisfied"
}

# Helper function
def plot_low_life_satisfaction(data, group_col, labels, title, xlabel, filename):
    temp = data.copy()
    temp["LowLifeSatisfaction"] = temp["LSMDVSWL"].isin([4, 5]).astype(int)

    summary = (
        temp.groupby(group_col)["LowLifeSatisfaction"]
        .mean()
        .reset_index()
    )

    summary["LowLifeSatisfactionPercent"] = summary["LowLifeSatisfaction"] * 100
    summary["Category"] = summary[group_col].map(labels)

    plt.figure(figsize=(9, 5))
    plt.bar(summary["Category"], summary["LowLifeSatisfactionPercent"])

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel("Low Life Satisfaction (%)")
    plt.xticks(rotation=30, ha="right")
    plt.tight_layout()

    plt.savefig(f"../outputs/figures/{filename}", dpi=300)
    plt.show()

    return summary
```

```python
# Life Satisfaction Distribution

life_counts = df_eda["LSMDVSWL"].value_counts().sort_index()
life_categories = [life_labels[i] for i in life_counts.index]

plt.figure(figsize=(8, 5))
plt.bar(life_categories, life_counts.values)

plt.title("Life Satisfaction Distribution Among Women")
plt.xlabel("Life Satisfaction Category")
plt.ylabel("Count")
plt.xticks(rotation=25, ha="right")
plt.tight_layout()

plt.savefig("../outputs/figures/life_satisfaction_distribution.png", dpi=300)
plt.show()
```

## Statistical Significance Testing

Chi-square tests of independence and Cramér's V were used to assess whether observed differences in low life satisfaction across categorical variables were statistically significant and to evaluate the strength of association.

```python
# ============================================================
# Chi-square Tests and Cramér's V
# ============================================================

from scipy.stats import chi2_contingency
import numpy as np
import pandas as pd

def chi_square_cramers_v(data, predictor, outcome="LowLifeSatisfaction"):
    """
    Performs Chi-square test of independence and calculates Cramér's V
    for association strength between a categorical predictor and outcome.
    """
    
    # Create contingency table
    table = pd.crosstab(data[predictor], data[outcome])
    
    # Run Chi-square test
    chi2, p_value, dof, expected = chi2_contingency(table)
    
    # Calculate Cramér's V
    n = table.to_numpy().sum()
    rows, cols = table.shape
    cramers_v = np.sqrt(chi2 / (n * (min(rows - 1, cols - 1))))
    
    return {
        "Variable": predictor,
        "Chi-square": chi2,
        "Degrees of freedom": dof,
        "p-value": p_value,
        "Cramer's V": cramers_v
    }

# Variables to test
variables_to_test = [
    "INCDGHH",    # Household income
    "FSCDVHF2",   # Food security
    "EDDVH3",     # Education
    "LBFDGWSS",   # Employment status
    "WDMDGDIF",   # Functional difficulty
    "CCCDGSKL",   # Musculoskeletal chronic condition
    "GEOGPRV"     # Province
]

# Run tests
chi_square_results = pd.DataFrame([
    chi_square_cramers_v(df_eda, var) for var in variables_to_test
])

# Format results
chi_square_results["Chi-square"] = chi_square_results["Chi-square"].round(3)
chi_square_results["p-value"] = chi_square_results["p-value"].apply(lambda x: "<0.001" if x < 0.001 else round(x, 4))
chi_square_results["Cramer's V"] = chi_square_results["Cramer's V"].round(3)

chi_square_results
```

```python
# ============================================================
# Save Chi-square and Cramér's V Results
# ============================================================

tables_path = "../outputs/tables"
os.makedirs(tables_path, exist_ok=True)

chi_square_results.to_csv(
    os.path.join(tables_path, "chi_square_cramers_v_results.csv"),
    index=False
)

chi_square_results
```
