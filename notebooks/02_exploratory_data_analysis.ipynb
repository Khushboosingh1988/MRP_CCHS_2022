```python
# 1. Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

pd.set_option("display.max_columns", None)

os.makedirs("../outputs/tables", exist_ok=True)
os.makedirs("../outputs/figures", exist_ok=True)

print("Libraries loaded")
```

```python
# 2. Load selected variables dataset

df_selected = pd.read_csv(
    "../data/processed/cchs_selected_variables.csv"
)

print("Selected dataset shape:", df_selected.shape)
df_selected.head()
```

```python
# 3. Create broader EDA dataset

df_eda = df_selected.copy()

df_eda = df_eda[df_eda["LSMDVSWL"].isin([1, 2, 3, 4, 5])]
df_eda = df_eda[df_eda["INCDGHH"] != 9]
df_eda = df_eda[df_eda["EDDVH3"] != 9]
df_eda = df_eda[df_eda["FSCDVHF2"] != 9]
df_eda = df_eda[~df_eda["WDMDGDIF"].isin([6, 9])]
df_eda = df_eda[~df_eda["CCCDGSKL"].isin([6, 9])]
df_eda = df_eda[df_eda["SDCDGIMM"] != 9]
df_eda = df_eda[df_eda["SDCDVFLA"] != 9]

print("EDA dataset shape:", df_eda.shape)
```

```python
# 4. Save EDA dataset

df_eda.to_csv(
    "../data/processed/cchs_2022_women_eda_dataset.csv",
    index=False
)

print("EDA dataset saved.")
```

```python
# ============================================================
# Final EDA Visualizations
# ============================================================

import os
os.makedirs("../outputs/figures", exist_ok=True)

life_labels = {
    1: "Very Satisfied",
    2: "Satisfied",
    3: "Neither",
    4: "Dissatisfied",
    5: "Very Dissatisfied"
}

# Helper function
def plot_low_life_satisfaction(data, group_col, labels, title, xlabel, filename):
    temp = data.copy()
    temp["LowLifeSatisfaction"] = temp["LSMDVSWL"].isin([4, 5]).astype(int)

    summary = (
        temp.groupby(group_col)["LowLifeSatisfaction"]
        .mean()
        .reset_index()
    )

    summary["LowLifeSatisfactionPercent"] = summary["LowLifeSatisfaction"] * 100
    summary["Category"] = summary[group_col].map(labels)

    plt.figure(figsize=(9, 5))
    plt.bar(summary["Category"], summary["LowLifeSatisfactionPercent"])

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel("Low Life Satisfaction (%)")
    plt.xticks(rotation=30, ha="right")
    plt.tight_layout()

    plt.savefig(f"../outputs/figures/{filename}", dpi=300)
    plt.show()

    return summary
```

```python
# Life Satisfaction Distribution

life_counts = df_eda["LSMDVSWL"].value_counts().sort_index()
life_categories = [life_labels[i] for i in life_counts.index]

plt.figure(figsize=(8, 5))
plt.bar(life_categories, life_counts.values)

plt.title("Life Satisfaction Distribution Among Women")
plt.xlabel("Life Satisfaction Category")
plt.ylabel("Count")
plt.xticks(rotation=25, ha="right")
plt.tight_layout()

plt.savefig("../outputs/figures/life_satisfaction_distribution.png", dpi=300)
plt.show()
```

```python
# Employment Status and Low Life Satisfaction

work_labels = {
    1: "Worked Last Week",
    2: "Did Not Work Last Week",
    6: "Not Applicable",
    9: "Not Stated"
}

employment_summary = plot_low_life_satisfaction(
    df_eda,
    "LBFDGWSS",
    work_labels,
    "Low Life Satisfaction by Employment Status",
    "Employment Status",
    "low_life_satisfaction_by_employment_status.png"
)

employment_summary
```
